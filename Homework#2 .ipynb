```python
def supergloss(s):
    definition='Word Definition:'+ s.definition();
    for hyponyms in s.hyponyms():
        definition= definition + '\n'+ 'Hyponyms:' + hyponyms.definition()
    for hypernyms in s.hypernyms():
        definition= definition + 'Hypernyms:' + hypernyms.definition()+ '\n'
    return definition

print(supergloss(wn.synset('water.n.00')))
```

```python
import nltk
from nltk.corpus import wordnet as wn

polysemy_nouns = list(wn.all_synsets('n'))

#compute number of words
numberOfWords = []
for i in polysemy_nouns:
    numberOfWords += i.lemma_names()
numberOfWords = set(numberOfWords)
wordsNumber = len(numberOfWords)

#compute total senses
senses = 0
for l in polysemy_nouns:
    senses = senses + 1

#compute averge 
avg = senses/wordsNumber

print('number of words:', wordsNumber)
print('senses:', senses)
print('average polysemy of nouns:', avg)
```
